## Data Cleaning and Preprocessing

```python
# Drop 'PassengerId' column as it's not required
df.drop(columns='PassengerId', inplace=True)
```

**Handle Missing Data**


- **Remove Rows or Columns:**

*Remove Rows with null values:*

We might consider dropping  rows with missing values in `Embarked` column because a relatively small number of rows have null values and we have enough data. So it doesn't significantly impact your dataset size or analysis.

Although it's categorical data (port names), so imputation i.e fill missing values with the mode (most frequent value) is also a straightforward approach. 


```python
# Drop rows with null values
df_clean =df.dropna(subset='Embarked')  
df_clean.head()
```

*Remove Column with null values:*

The `Cabin` column has a large number of missing values, which can make direct imputation challenging. However, the presence or absence of Cabin data itself might be informative

But `Cabin` data is not deemed critical for my analysis so I consider dropping the column entirely.


```python
df_clean = df_clean.drop(columns=['Cabin'])  # Drop a specific column
```

- **Impute or Fill Missing Values**

    ||Strategies / Approaches||:

    - Mean or median imputation: Fill missing values with the mean or median values of that column. This is a common approach when the distribution of column is relatively normal.

    - Predictive imputation: Use predictive models (e.g., regression models based on other features) to estimate missing values more accurately.


*Mean Imputation:*

`Age` is an important feature, but it has a significant number of missing values. Imputing these values requires careful consideration of the distribution and impact on our analysis.

Although this column has a large number of null values and is critical for your analysis, I choose to impute instead drop that column. 

```python
# Fill missing values with mean age of passengers
mean_age = df_clean['Age'].mean()
df_clean['Age'].fillna(mean_age, inplace=True)
```

> Keep in mind:
- Impute numerical values: For numerical columns, you can replace null values with the mean, median, or mode of that column.
- Impute categorical values: For categorical columns, replace null values with the most frequent category.


```python
# Check if there are any missing values left in the dataset
print(df_clean.isnull().sum())
```

###  1: Create a simple line plot using Matplotlib to display the trend of a list of values over time.


Here, we have to display trend of values over time.

So for this purpose, we'll use the `"Flights Dataset"`which contains information about the number of passengers flying with a specific airline each month, making it a good candidate for demonstrating a "time series plot".

```python
# Load the Air Passengers dataset from Seaborn
df2 = sns.load_dataset('flights')
# df.head()
# df.info()


# Convert the 'year' and 'month' columns to datetime format:
    # Convert 'year' and 'month' to strings
df2['year'] = df2['year'].astype(str)
df2['month'] = df2['month'].astype(str)
    # Create 'date' column by combining 'year' and 'month' as strings
df2['date'] = pd.to_datetime(df2['year'] + '-' + df2['month'], format='%Y-%b')    # format string specifies: year (%Y)-4digits(1949), month (%b)-three-letter abbr(Jan, Feb,..)
# Verify changes
df2.info()
```

```python
# Validate dateformat
try:
    pd.to_datetime(df2['date'], format='%d-%b-%Y', errors='raise')
    print('Date formatted!')
except ValueError:
    pass
```

```python
# Sort the dataframe based on date
df2.sort_values(by='date',inplace=True)
```

```python
                                            ## Line Plot
    
# Create a line plot of passengers over time (using plt)
plt.figure(figsize=(10, 6))

# Set the background color of the figure
plt.figure(facecolor='lightblue')
# plt.plot(df['date'], df['passengers'],)
plt.plot(df2['date'], df2['passengers'], marker='o',linestyle='-', color='b', label='Values', markersize=3)



plt.title('Airline Passengers Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()

# add gridlines for better readability
plt.grid(True)
# improve spacing between subplots
plt.tight_layout()
# Display the plot
plt.show()
```

### 2. Plot a bar chart using Matplotlib to show the frequency of different categories in a dataset.

We can plot frequency of following categories: # Sex,Embarked,Pclass 


```python
                                      ## Bar Chart
    
# List of columns for which we want to plot bar-plot / value counts
columns_to_plot = ['Sex', 'Pclass', 'Embarked']

# Plotting bar charts for each column in columns_to_plot
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))  # Adjusted to one row with three columns

for i, column in enumerate(columns_to_plot):
    # Calculating Row and Column Indices
    row = 0  # Since we want all plots in the same row
    col = i  # Column index matches the iteration index i
    # Otherwise
    # row = i // 2
    # col = i % 2
    
    ax = axes[col]  # Accessing the correct subplot
    
    # Plotting the bar chart for value counts
    df[column].value_counts().plot(kind='bar', ax=ax, color='green')
    
    # Adding titles and labels
    ax.set_title(f'Value Counts of {column}')
    ax.set_xlabel(column)
    ax.set_ylabel('Count')

    # Rotate x-axis labels for better readability
    ax.tick_params(axis='x', rotation=45)  # Adjusting x-axis tick labels
    
# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

### 6: Plot a heatmap using Seaborn to visualize the correlation matrix of a dataset

```python
                                    ## Correlation Matrix Heatmap
    
# Compute correlation matrix
cm = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Blues', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### 8: Customize the appearance of a Seaborn plot by changing the color palette, adding titles, and modifying axis labels.


Although Seaborn creates attractive plots, we don't necessarily need to customize every aspect like we do with Matplotlib. However, I also included some Matplotlib code for more complex visualizations, which I later commented out.

```python
                                    ### Complex plot using Seaborn ###
# Import Seaborn for visualization
import seaborn as sns
# Import Matplotlib for plotting
import matplotlib.pyplot as plt

# Set Seaborn style and color palette (Optionl)
# sns.set(style="whitegrid", palette="pastel")   
sns.set(style='whitegrid')    # Choose a light style like 'whitegrid', 'ticks', 'white', etc.
blue_palette = sns.color_palette('Blues')  # Choose a blue color palette
sns.set_palette(blue_palette)  # Set the chosen palette

# Choose features to plot
x_variable,y_variable,category_variable,size_variable,style_variable='Age','Fare','Sex', 'Pclass','Survived'

# Set figsize
# plt.figure(figsize=(10, 8))

# Create a plot (any):
# Creating plot using relplot for complex visualization
sns.relplot(
    x=x_variable,         # Column name from df for x-axis
    y=y_variable,         # Column name from df for y-axis
    hue=category_variable,   # Column name from df for coloring points
    size=size_variable,   # Column name from df for sizing points
    style=style_variable, # Column name from df for styling points
    data=df,                # DataFrame containing the data
    palette='Set2',         # Color palette to use for hue variable
    sizes=(20, 200),        # Size range for points
    alpha=0.8,               # Transparency of points
    height=8,                # Height of the plot in inches
    legend='brief'  # Ensure only one legend is shown for categorical variables
)

# Adding titles and labels
plt.title('Complex Visualization using Seaborn - Titanic Dataset', fontsize=14, fontweight='bold', y=1.1, pad=10)
# plt.xlabel('Passenger Age', fontsize=14, fontweight='bold', labelpad=15)
# plt.ylabel('Fare Paid', fontsize=14, fontweight='bold', labelpad=15, ha='center', va='center')

# Adding a legend (Customized)
# plt.legend(title='Categories:', loc='upper right', bbox_to_anchor=(1.2, 1))

# Show plot
plt.show()


# Parameters Brief:
    # data=df
    # x,y,style, size,hue = column_name
    # sizes=(min_size, max_size) 
    # alpha=transparency_value (or any value between 0 and 1)
    # palette='palette_name' or palette=color_palette
    # legend= brief, 
    # ha, va, loc, pad, labelpad, rotation, ----> for desired positioning and appearance of your plot titles and labels.
    # loc, y, labelpad,---> fine-tune the positioning and appearance of plot titles and labels
    # bbox_to_anchor=(1.2, 1) parameter ---> used for adjusting the position of legends or other elements in the plot, not for titles, xlabels, or ylabels.
```

> **Note:** sns.relplot() creates a FacetGrid object by default, which manages facets (subplots) based on categorical variables like hue, col, or row. When you call plt.legend() directly after sns.relplot(), it adds another legend to the same plot. So you can use either sns legend or cutomize with matplotlib.

### 9: Create a violin plot using Seaborn to visualize the distribution of a dataset across different categories.


```python
# Set Seaborn style and color palette
sns.set(style='whitegrid')

# Set figsize
plt.figure(figsize=(10, 8))

# Create violin plot
sns.violinplot(x='Pclass', y='Fare', data=df,hue=df['Sex'],palette='Set2')

# Adding titles and labels with adjusted alignment
plt.title('Distribution of Fare by Passenger Class and Sex (Violin Plot)', fontsize=14, fontweight='bold', ha='center', va='bottom')
plt.xlabel('Passenger Class', fontsize=12, fontweight='bold', ha='center', va='top')
plt.ylabel('Fare', fontsize=12, fontweight='bold', ha='right', va='center')

# Show plot
plt.show()
```

### 10: Combine Matplotlib and Seaborn to create a complex visualization, such as overlaying a KDE plot on a histogram.

```python
# Set Seaborn style and color palette
sns.set(style='white')

# Set figsize
plt.figure(figsize=(8, 4))

# Plotting overlapping histograms with KDE for 'Fare' column
sns.histplot(data=df[df['Pclass'] == 1]['Fare'], color='blue',kde=True, alpha=0.5, label='First Class')  
sns.histplot(data=df[df['Pclass'] == 2]['Fare'], color='orange', kde=True, alpha=0.5, label='Second Class')  

# Adding title,abels and legend
plt.title('Distribution of Fare Paid by Passengers Class - Overlaying', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Fare ($)', fontsize=12, fontweight='bold')
plt.ylabel('Frequency / Density', fontsize=12, fontweight='bold')
plt.legend()

# Showing the plot
plt.show()
```
