###  Dataset Intro:

##### The Boston Housing Dataset

The Boston Housing Dataset is a derived from information collected by the U.S. Census Service concerning housing in the area of [Boston MA](https://www.cs.toronto.edu/~delve/data/boston/bostonDetail.html "Data DESCR")

Let's calculate the percentage of outliers in each column:

```python
## Using IQR Method:

# Iterate over each column `k` and its corresponding data `v` in the `data` dictionary or DataFrame
for k, v in boston_df.items():
    # Calculate the first quartile (Q1) and third quartile (Q3) of the data `v`
    q1 = v.quantile(0.25)
    q3 = v.quantile(0.75)
    
    # Calculate the Interquartile Range (IQR)
    iqr = q3 - q1
    
    fence_low = q1 - (1.5*iqr)
    fence_high = q3 + (1.5*iqr)
    
    # Identify outliers using the 1.5*IQR rule
    v_col = v[(v <= q1 - 1.5 * irq) | (v >= q3 + 1.5 * irq)]
    
    # Calculate the percentage of outliers in the column `k`
    outlier_perc = np.shape(v_col)[0] * 100.0 / np.shape(boston_df)[0]
    
    # Print the results
    # Format the output string using f-string
    formatted_output = f"Column '{k}' outliers: {outlier_perc:.2f}%"
    print(formatted_output)
```

###  Data Preparation
Preparing the data for the model training:


```python
# Prepare your data as a pandas DataFrame or NumPy array.
# Accessing the i/o variables

# Features (input variables)
# X = boston_df_f.drop(columns='MEDV')    # all variables except MEDV
X =boston_df_f[['LSTAT', 'INDUS', 'NOX', 'PTRATIO', 'RM', 'TAX', 'DIS', 'AGE']]   # select specific variables
# Target (output variable)
y = boston_df_f['MEDV']  

print("\nInput Variables:\n",X.head())
print("\nOutput Variables:\n",X.head())
```

### Splitting the data into training and testing¬†sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=5)

# Parameters:
    # X and y: The input features and output labels.
    # test_size=0.2: Specifies that 20% of the data should be used for testing (i.e., an 80/20 split).
    # random_state=42: Sets the random seed for reproducibility. It ensures that the split is deterministic, meaning each time you run the code, you'll get the same split.

# Outputs:
    # X_train and X_test: These are the feature sets for training and testing, respectively.
    # y_train and y_test: These are the corresponding labels for training and testing.
```

```python
print('Training Data:')
print('X',X_train.shape)
print('y',y_train.shape)
print('')
print('Testing Data:')
print('X',X_test.shape)
print('y',y_test.shape)
```

### Train a Linear Regression model.

#### Create a Linear Regression Model

```python
# Initialize the model
# Create a LinearRegression object from sklearn
lr = linear_model.LinearRegression()
```

####  Fit the Model

```python
# Fit the Model to the data
lr_model = lr.fit(X_train,y_train)
```

#### Print the model's coefficients and intercept.


```python
# Print coefficients and intercept
print('Coefficients:', lr_model.coef_)
print('')
print('Intercept:', lr_model.intercept_)
```

```python
# Converting the coefficient values to a dataframe
coef_df = pd.DataFrame([X_train.columns,lr_model.coef_]).T
coef_df = coef_df.rename(columns={0: 'Attribute', 1: 'Coefficients'})
coef_df
```

**Interpreting the coefficients:**

These interpretations provide insights into how each predictor variable (attribute) impacts the target variable when all other variables are held constant.

Based on the provided coefficients, we can interpret:

- Negative Coefficients (LSTAT, INDUS, NOX, PTRATIO, TAX, DIS, AGE) indicate factors that tend to lower housing prices when they increase.
- Positive Coefficient (RM) indicates a factor that tends to increase housing prices when it increases.
- Magnitude of the coefficient indicates the strength of the effect on housing prices. For example, NOX has a large negative impact, while RM has a significant positive impact.

```python
# Predict on the test set
y_test_pred = lr_model.predict(X_test)
y_test_pred # y_pred
```

Evaluate the model

**Outlier Analysis**

```python
# Function to detect outliers in every feature

def detect_outliers(df):
    # Get list of columns in the dataframe
    columns = list(df)
    
    # Initialize an empty dataframe to store outliers information
    outliers_info = pd.DataFrame(columns=['Feature', 'Number of Outliers'])
    
    # Iterate through each column in the dataframe
    for col in columns:
        # Check if the column contains numerical data
        if col in df.select_dtypes(include=np.number).columns:
            # Calculate the first quartile (Q1) and third quartile (Q3)
            q1 = df[col].quantile(0.25)
            q3 = df[col].quantile(0.75)
            
            # Calculate the interquartile range (IQR)
            iqr = q3 - q1
            
            # Calculate the lower and upper bounds for outlier detection
            fence_low = q1 - (1.5 * iqr)
            fence_high = q3 + (1.5 * iqr)
            
            # Count the number of outliers for the current feature
            num_outliers = [df.loc[(df[col] < fence_low) | (df[col] > fence_high)].shape[0]]

            # Append the feature and number of outliers to the outliers_info dataframe
            outliers_info = pd.concat([outliers_info, pd.DataFrame({'Feature': col, 'Number of Outliers': num_outliers})], ignore_index=True)
    
    return outliers_info

# usage:
outliers_info = detect_outliers(iris_df)
outliers_info
```

```python
# % of Outliers:
for k, v in iris_df.items():
    if v.dtype in [np.int64, np.float64]:  # Check if column is numeric
        q1 = v.quantile(0.25)
        q3 = v.quantile(0.75)
        irq = q3 - q1
        v_col = v[(v <= q1 - 1.5 * irq) | (v >= q3 + 1.5 * irq)]
        perc = np.shape(v_col)[0] * 100.0 / np.shape(iris_df)[0]
        # String Formatting with % Operator: (placeholder, value)
        print("Column %s outliers = %.2f%%" % (k, perc))
        
#
# Print outliers
# print("Indices of outliers:")
# print(np.where(outliers))
```

The absence of outliers (0.00% outliers) in both the petal length and petal width columns suggests that these features in the Iris dataset are well-behaved and do not exhibit any extreme values or deviations from the majority of the data.

A few were seen  sepal width, but for this task we're not excluding these.

`Note`
It's important to note that the absence of outliers in this context is based on the specific threshold of 1.5 times the IQR (Interquartile Range). Adjusting this threshold might reveal different results or a different interpretation of outliers.
Having a clean and well-distributed dataset without outliers can lead to more reliable and accurate model predictions.
Always consider domain knowledge and the specific context of your analysis when interpreting outlier detection results.
