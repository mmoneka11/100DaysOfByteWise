**1. Recursive Factorial**
   - Write a recursive function to calculate the factorial of a given number.
   - Expected output: If the input is 5, the output should be "The factorial of 5 is 120."
   
`Note:`
*Factorial* is the product of positive numbers, each number is one less than the preceding number. It is denoted by n!.

```python
# Program:

def recursive_factorial(n):
    # base case
    if n==0:
        return 1
    # recursively calls itself with n-1 untill it reaches base case, multiplying each value of 'n'
    return n*recursive_factorial(n-1)

recursive_factorial(5)
```

**2. Palindrome Linked List**
   - Write a program to determine if a given linked list is a palindrome.
   - Expected output: If the linked list is `1 -> 2 -> 3 -> 2 -> 1`, the output should be "The linked list is a palindrome." If the linked list is `1 -> 2 -> 3 -> 4 -> 5`, the output should be "The linked list is not a palindrome."

*To create a linked list, follow these steps:*

Define a Node class: This class represents individual nodes of the linked list. Each node typically contains two components: the data value and a reference to the next node in the list.

Instantiate Node objects: Create instances of the Node class for each element you want to add to the linked list. Assign values to the data component of each node.

Link the nodes: Set the next attribute of each node to point to the next node in the list. This creates the linkage between nodes, forming the linked list.

Optional: Define helper functions: You can define helper functions to perform common operations on the linked list, such as printing the list or adding/removing nodes.


*To determine if a linked list is a palindrome, follow these steps:*

Reverse the Second Half of the Linked List:
Traverse the linked list to find the middle node. Reverse the second half of the linked list.

Compare the First Half with the Reversed Second Half:
Traverse the first half of the original linked list and the reversed second half simultaneously, comparing each node's value.

Check Palindrome:
If all nodes match, then the linked list is a palindrome. Otherwise, it's not.

```python
# Program:

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
    # Define a function to print the linked list
    @staticmethod
    def print_linked_list(head):
        while head:
            print(head.val, end=" -> ")
            head = head.next
        print("None")

def is_palindrome(head):
    values = []
    while head:
        values.append(head.val)
        head = head.next
    return values == values[::-1]

# usage:
head1 = ListNode(1, ListNode(2, ListNode(3, ListNode(2, ListNode(1)))))
head2 = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))

print("Linked list 1:")
ListNode.print_linked_list(head1)
print("The linked list is a palindrome." if is_palindrome(head1) else "The linked list is not a palindrome.")

print("\nLinked list 2:")
ListNode.print_linked_list(head2)
print("The linked list is a palindrome." if is_palindrome(head2) else "The linked list is not a palindrome.")
```

**3. Merge Sorted Arrays**
   - Write a function that takes two sorted arrays and merges them into a single sorted array.

- Expected output: If the two input arrays are `[1, 3, 5]` and `[2, 4, 6]`, the output should be `[1, 2, 3, 4, 5, 6]`.

`Steps`

Initialize empty list: Create an empty list to store the merged sorted array.

Initialize pointers: Initialize two pointers, one for each input array, to track the current position while merging.

Compare elements: Compare elements from both input arrays. Append the smaller element to the merged array and move the corresponding pointer forward.

Handle remaining elements: Continue this process until one of the input arrays is completely traversed. If any elements remain in either array, append them to the merged array.

Return merged array: Once both input arrays are fully traversed, return the merged array.


```python
# Program:

def merge_sorted_arrays(arr1, arr2):
    merged = []
    i = j = 0

    # Iterate through both arrays simultaneously
    while i < len(arr1) and j < len(arr2):
        # Compare elements from both arrays
        if arr1[i] < arr2[j]:
            merged.append(arr1[i])
            i += 1
        else:
            merged.append(arr2[j])
            j += 1

    # Add remaining elements from arr1, if any
    merged.extend(arr1[i:])

    # Add remaining elements from arr2, if any
    merged.extend(arr2[j:])

    return merged

# Test
arr1 = [1, 3, 5]
arr2 = [2, 4, 6]
print(merge_sorted_arrays(arr1, arr2))  # Output: [1, 2, 3, 4, 5, 6]


# We initialize the merged list as an empty list.
# We use two pointers (i and j) to iterate through both arrays simultaneously.
# We compare elements at the current indices of both arrays and append the smaller one to the merged list.
# After one array is exhausted, we extend the merged list with the remaining elements of the other array.
# Finally, we return the merged sorted array
```

**4. Binary Search Tree**
   - Implement a Binary Search Tree (BST) data structure, including methods for insertion, deletion, and search.
   - Expected output: The program should be able to perform various BST operations and print the results.


`Steps:`

Define Node Class: Create a class for the nodes of the BST. Each node should have a value, a reference to its left child, and a reference to its right child.

Initialize BST Class: Create a class for the Binary Search Tree. Initialize it with a root node, initially set to None.

Insertion Method: Implement a method to insert a new node into the BST. This method should follow the rules of a BST: smaller values to the left and larger values to the right.

Search Method: Implement a method to search for a value in the BST. This method should recursively traverse the tree, comparing values until the target value is found or the traversal reaches a leaf node.

Deletion Method: Implement a method to delete a node from the BST. Deleting a node requires handling different cases: a node with no children, a node with one child, and a node with two children.

Case 1: Node with No Children: If the node to be deleted has no children, simply remove it from the tree by setting its parent's reference to None.

Case 2: Node with One Child: If the node to be deleted has only one child, replace the node with its child.

Case 3: Node with Two Children: If the node to be deleted has two children, find the inorder successor (the smallest node in the right subtree), replace the node's value with the inorder successor's value, and then delete the inorder successor.

Testing: Test the BST operations by inserting nodes, searching for values, and deleting nodes.



```python
# Program:

class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def insert(self, value):
        if self.root is None:
            self.root = TreeNode(value)
        else:
            self._insert_recursive(self.root, value)
    
    def _insert_recursive(self, node, value):
        if value < node.value:
            if node.left is None:
                node.left = TreeNode(value)
            else:
                self._insert_recursive(node.left, value)
        else:
            if node.right is None:
                node.right = TreeNode(value)
            else:
                self._insert_recursive(node.right, value)

    def print_bst(self):
        self._print_inorder(self.root)
        print()

    def _print_inorder(self, node):
        if node:
            self._print_inorder(node.left)
            print(node.value, end=" ")
            self._print_inorder(node.right)
        
    def delete(self, value):
        self.root = self._delete_recursive(self.root, value)

    def _delete_recursive(self, node, value):
        if node is None:
            return node
        
        if value < node.value:
            node.left = self._delete_recursive(node.left, value)
        elif value > node.value:
            node.right = self._delete_recursive(node.right, value)
        else:
            if node.left is None:
                return node.right
            elif node.right is None:
                return node.left
            else:
                successor = self._find_min(node.right)
                node.value = successor.value
                node.right = self._delete_recursive(node.right, successor.value)
        
        return node

    def _find_min(self, node):
        while node.left is not None:
            node = node.left
        return node

    def search(self, value):
        return self._search_recursive(self.root, value)

    def _search_recursive(self, node, value):
        if node is None or node.value == value:
            return node
        elif value < node.value:
            return self._search_recursive(node.left, value)
        else:
            return self._search_recursive(node.right, value)


# Example usage
if __name__ == '__main__':
    bst = BST()

    # Insertion
    bst.insert(50)
    bst.insert(30)
    bst.insert(20)
    bst.insert(40)
    bst.insert(70)
    bst.insert(60)
    bst.insert(80)
    
     # Print BST
    print("Binary Search Tree:")
    bst.print_bst()
    
    # Search
    search_value = 70
    result = bst.search(search_value)
    if result:
        print(f"{search_value} found in the BST.")
    else:
        print(f"{search_value} not found in the BST.")

    # Deletion
    bst.delete(20)
    bst.delete(30)
    bst.delete(50)

    # Search after deletion
    search_value = 70
    result = bst.search(search_value)
    if result:
        print(f"{search_value} found in the BST.")
    else:
        print(f"{search_value} not found in the BST.")
```

**5. Longest Palindromic Substring**
   - Write a program to find the longest palindromic substring within a given string.
   - Expected output: If the input string is "babad", the output should be "bab" or "aba". If the input string is "cbbd", the output should be "bb".


```python
# Program:

def longest_palindromic_substring(s):
    if not s:
        return ""


    def expand_around_center(left, right):
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        return s[left+1:right]


    longest = ""
    for i in range(len(s)):
        # Check for odd-length palindromes (single character center)
        odd_palindrome = expand_around_center(i, i)
        if len(odd_palindrome) > len(longest):
            longest = odd_palindrome

        # Check for even-length palindromes (pair of characters center)
        even_palindrome = expand_around_center(i, i + 1)
        if len(even_palindrome) > len(longest):
            longest = even_palindrome

    return longest

# Usage:
input_string1 = "babad"
output1 = longest_palindromic_substring(input_string1)
print(f"Longest palindromic substring in '{input_string1}' is '{output1}'")

input_string2 = "cbbd"
output2 = longest_palindromic_substring(input_string2)
print(f"Longest palindromic substring in '{input_string2}' is '{output2}'")


# This function uses dynamic programming to build a table dp where dp[i][j] indicates whether the substring from index i to index j is a palindrome. It then iterates through all substrings to find the longest palindromic substring and returns it
#dynamic programming is a technique commonly used to solve optimization problems by breaking them down into simpler subproblems and solving each subproblem only once, storing its result for future reference. In the case of finding the longest palindromic substring, dynamic programming helps avoid redundant computations by storing the results of subproblems in a table.

# In this specific problem:
# We initialize a table dp where dp[i][j] indicates whether the substring from index i to index j is a palindrome.
# We fill in the base cases: single characters are considered palindromes, and we check for palindromes of length 2.
# We iterate through all substrings of length greater than 2, checking if the substring is a palindrome by comparing characters and referring to the results of smaller substrings.
# We update the start and max_len variables whenever we find a longer palindrome.
```

**6. Merge Intervals**
   - Write a program to merge overlapping intervals in a list of intervals.
   - Expected output: If the input is `[(1, 3), (2, 6), (8, 10), (15, 18)]`, the output should be `[(1, 6), (8, 10), (15, 18)]`.


```python
# Given an array of intervals with start & end value i.e intervals[i]=[start,end]
# For each of these intervals, take all the overlapping intervals then merge them to give a new output array of intervals.
# Since we merge overlapping intervals the result ---->non-overlapping intervals
# e.g:
    # [(1,4),(4,5)]  ---> these are overlapping like 1 to 4 & 4 to 5: 1<--->4<--->5
    # Output: [1,5]
    
```

```python
# Program:
  
class Solution:
    
    def merge_intervals(self,intervals):
    # -->O(nlogn)  # c/z O(nlog)+O(n)
        if intervals:

            # sorting entire input 
            # sort a list of pairs, sort by start value using first value interval[0], 
            intervals.sort(key= lambda interval:interval[0]) # interval[1] for end value
            # declare a var to put merged intervals, a list is initialized here
            merged= [intervals[0]]  # first value given alrdy


            # iterating through input
            # iterate through every single interval in sorted order
            # instead of having interval itself we use start & end value , extract them 
            for start, end in intervals[1:]:     # we can skip first one c/z we alrdy added first interval in output
             # chk interval is overlapping with most recent interval 
                last_end= merged[-1][1]# get most recent added interval & get end value of it, get the end value of it
                # chk current interval is less than end
                if  start<= last_end:  # overlapping
#                if not merged or start<= last_end:   # to chk in case if initialized list empty 
                    # merge them
                    merged[-1][1]=max(last_end, end)  # max of last end and current end            
                else: # overlapping
                    merged.append([start,end])
            return merged
        
        else:
            return []


sol=Solution()

intervals= [[1,3], [2,6], [8,10], [15,18]]
merged_intervals=sol.merge_intervals(intervals)
print(f'Intervals: {intervals}')
print(f'Merged Intervals: {merged_intervals}')


# Reason of taking max():-
# if we have intervals like:
# [1,5], [2,4] ---> first one has larger end value than second one
# b/c we'd iterate through them in this order b/c firs has starting value of 1 and second has starting value of 2,
# but larger end value of first than second one
# if we just take 4 and merge this together and set [1,4], [2,4] --> we're actually making interval smaller
# we want to keep 5 b/c merging these together is gonna result in [1,5]
```

**7. Maximum Subarray**
   - Write a program to find the maximum sum of a contiguous subarray within a given array.
   - Expected output: If the input array is `[-2, 1, -3, 4, -1, 2, 1, -5, 4]`, the output should be `6`, as the maximum subarray is `[4, -1, 2, 1]`.



```python
# Program:

class MaxSubArray:
#     def max_subarray(self, nums:List[int])->int:
    def max_subarray(self, nums):
        # initialize max subarray value to first value in array,
        # that's like the default val we give it c/z we've to give it something and it can't be zero,
        # b/c we know we've negative values in array
        maxSub= nums[0] 
        # initialize current sum to 0 c/z we're gonna be computing it constantly
        curSum= 0

        # iterate through each number in nums
        for n in nums:
            # if we've -ve prefix --> remove that portion from current sum
            # curSum at any point is negative if it's less than 0, so we're gona reset it back to 0
            if curSum<0:
                curSum=0
            # add current num to this
            curSum+=n # This will make sure we're always computing max that we can
            # now curSum could be the possible max so we're just gonna update our max subarray to max of itself to max tht we just computed
            maxSub=max(maxSub, curSum)

        # return whatever we computed
        return maxSub

    
sol=MaxSubArray()

input_array = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
max_sum = sol.max_subarray(input_array)
print(f"The maximum sum of a contiguous subarray is {max_sum}")
```


**8. Reverse Linked List**
   - Write a program to reverse a singly-linked list.
   - Expected output: If the input linked list is `1 -> 2 -> 3 -> 4 -> 5`, the output should be `5 -> 4 -> 3 -> 2 -> 1`.

```python
# Program:

class Node:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None
        
    # Create a linked list from a list of values i.e [1, 2, 3, 4, 5]    
    def create_linked_list(self, values):
        if not values:
            return
        
        self.head = Node(values[0])
        current = self.head
        
        for value in values[1:]:
            current.next = Node(value)
            current = current.next

    # Function to print the linked list
    def print_linked_list(self):
        current = self.head
        while current:
            print(current.value, end=" -> ")
            current = current.next
        print("None")

    # reverselist function
    def reverse_list(self):
        prev, curr = None, self.head
        
        while curr:
            nxt = curr.next  # Store the next node in the original list
            curr.next = prev  # Reverse the pointer to the previous node
            
            # Move pointers one step forward
            prev = curr
            curr = nxt
        
        # At the end of the loop, 'prev' will be the new head of the reversed list
        self.head = prev

# Usage:
values = [1, 2, 3, 4, 5]

# Create a linked list
linked_list = LinkedList()
linked_list.create_linked_list(values)

# Print the original linked list
print("Original Linked List:")
linked_list.print_linked_list()

# Reverse the linked list
linked_list.reverse_list()

# Print the reversed linked list
print("\nReversed Linked List:")
linked_list.print_linked_list()
```

**9. Minimum Edit Distance**
   - Write a program to calculate the minimum number of operations (insertions, deletions, or substitutions) required to transform one string into another.
   - Expected output: If the two input strings are "kitten" and "sitting", the output should be `3`.


```python
# Program:

def min_edit_distance(str1, str2):
    m = len(str1)
    n = len(str2)
    
    # Initialize a 2D list (dp) to store the minimum edit distance
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    # Initialize the base cases
    for i in range(m + 1):
        dp[i][0] = i  # Cost of deleting all characters of str1
    for j in range(n + 1):
        dp[0][j] = j  # Cost of inserting all characters of str2
    
    # Fill the dp table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if str1[i - 1] == str2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]  # Characters are the same, no operation needed
            else:
                dp[i][j] = 1 + min(dp[i - 1][j],    # Deletion
                                   dp[i][j - 1],    # Insertion
                                   dp[i - 1][j - 1])  # Substitution
    
    # The bottom-right cell contains the minimum edit distance
    return dp[m][n]

# Example usage
if __name__ == "__main__":
    str1 = "intention"
    str2 = "execution"
    
    # Calculate the minimum edit distance
    result = min_edit_distance(str1, str2)
    
    # Print the result
    print(f"The minimum number of operations required to transform '{str1}' into '{str2}' is {result}")
```

**10. Boggle Game**

   - Implement a program that solves the Boggle game, given a board and a list of words.
   - Expected output: The program should print all the words found in the Boggle board.


`NOTE:` Boggle Game is a word game can be played with 2 or more players. Goal is to find the words with in random combination of letters on the dice in the game tray.


```python
# Program:

class BoggleSolver:
    def __init__(self, dictionary):
        self.dictionary = set(dictionary)  # Store dictionary words in a set for O(1) lookup
        self.M = 0  # Initialize board dimensions
        self.N = 0  # Initialize board dimensions
        self.boggle = []  # Initialize boggle board

    def set_board(self, boggle):
        self.boggle = boggle
        self.M = len(boggle)
        self.N = len(boggle[0]) if self.M > 0 else 0

    def is_valid_word(self, word):
        return word in self.dictionary

    def find_words(self):
        if self.M == 0 or self.N == 0:
            return set()  # Return empty set if board is empty

        found_words = set()  # Set to store found words
        visited = [[False for _ in range(self.N)] for _ in range(self.M)]  # Initialize visited matrix
         
        # Recursive Word Search
        # Utility function to recursively find all possible words starting from position (i, j)
        def find_words_util(i, j, current_str):
            visited[i][j] = True
            current_str += self.boggle[i][j]

            if self.is_valid_word(current_str):
                found_words.add(current_str)

            # Check all 8 possible directions (up, down, left, right, and diagonal)
            for row in range(i - 1, i + 2):
                for col in range(j - 1, j + 2):
                    if 0 <= row < self.M and 0 <= col < self.N and not visited[row][col]:
                        find_words_util(row, col, current_str)

            visited[i][j] = False  # Backtrack

        # Iterate over each cell in the board and start finding words
        for i in range(self.M):
            for j in range(self.N):
                find_words_util(i, j, "")

        return found_words


#  usage:
if __name__ == "__main__":
    # Define the dictionary and boggle board
    dictionary = ["CAT", "HAT", "BAT", "SAT", "SAG", "EAT", "BEAT", "DOG", "GOT", "FROG"]
    boggle = [
        ["C", "A", "T", "H"],
        ["B", "G", "O", "A"],
        ["S", "A", "G", "L"],
        ["E", "F", "R", "D"]
    ]

    # Create instance of BoggleSolver
    solver = BoggleSolver(dictionary)

    # Set the boggle board
    solver.set_board(boggle)

    # Find and print the words present in the boggle board
    print("Following words of the dictionary are present:")
    found_words = solver.find_words()
    print(found_words)
```
