Let's calculate the percentage of outliers in each column:

```python
## Using IQR Method:

# Iterate over each column `k` and its corresponding data `v` in the `data` dictionary or DataFrame
for k, v in boston_df.items():
    # Calculate the first quartile (Q1) and third quartile (Q3) of the data `v`
    q1 = v.quantile(0.25)
    q3 = v.quantile(0.75)
    
    # Calculate the Interquartile Range (IQR)
    iqr = q3 - q1
    
    fence_low = q1 - (1.5*iqr)
    fence_high = q3 + (1.5*iqr)
    
    # Identify outliers using the 1.5*IQR rule
    v_col = v[(v <= q1 - 1.5 * iqr) | (v >= q3 + 1.5 * iqr)]
    
    # Calculate the percentage of outliers in the column `k`
    outlier_perc = np.shape(v_col)[0] * 100.0 / np.shape(boston_df)[0]
    
    # Print the results
    # Format the output string using f-string
    formatted_output = f"Column '{k}' outliers: {outlier_perc:.2f}%"
    print(formatted_output)
```

###  Data Preparation
Preparing the data for the model training:


```python
# Prepare your data as a pandas DataFrame or NumPy array.
# Accessing the i/o variables

# Features (input variables)
# X = boston_df_f.drop(columns='MEDV')    # all variables except MEDV
X =boston_df_f[['LSTAT', 'INDUS', 'NOX', 'PTRATIO', 'RM', 'TAX', 'DIS', 'AGE']]   # select specific variables
# Target (output variable)
y = boston_df_f['MEDV']  

print("\nInput Variables:\n",X.head())
print("\nOutput Variables:\n",X.head())
```

#### Interpreting the coefficients
These interpretations provide insights into how each predictor variable (attribute) impacts the target variable when all other variables are held constant.


```python
print("\nBased on the provided coefficients, we can interpret:\n")

for index, row in coeff_df.iterrows():
    feature_name = row['Feature']
    coefficient_value = row['Coefficient']
    
    print(f"Feature '{feature_name}' has a coefficient of {coefficient_value:.2f}:")
    
    if coefficient_value > 0:
        print(f"\t- A positive coefficient suggests that an increase in '{feature_name}' is associated with an increase in the target variable.")
    elif coefficient_value < 0:
        print(f"\t- A negative coefficient suggests that an increase in '{feature_name}' is associated with a decrease in the target variable.")
    else:
        print(f"\t- The coefficient suggests no significant impact of '{feature_name}' on the target variable.")
        
    print(f"\t- The magnitude of {coefficient_value:.2f} indicates the strength of this relationship.")
    print()  # blank line
    print('\t')
    
    # Sort coefficients by absolute value to identify the strongest impact
    coeff_df_sorted = coeff_df.reindex(coeff_df['Coefficient'].abs().sort_values(ascending=False).index)

    # Highlight the strongest coefficient
    if index == coeff_df_sorted.index[0]:
        print(f"Feature '{feature_name}' has the strongest impact on the target variable based on its coefficient of {coefficient_value:.2f}.")
        print()  # blank line for separation
```

###  Evaluation with Cross Validation

(with all features involved)

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split, cross_val_score

# Loading data
# datafile = "Boston Housing/housing.data"
df =boston_df_f

# Splitting X and y
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
# X
 

# # Defining regressors i.e Create instances of the regressors
lr = LinearRegression()
rf = RandomForestRegressor()
svr = SVR()

# List of regressors and their names
regressors = [lr,rf,svr]
regressor_names = ["LR","RF","SVC"]

# Spliting data into training and testing sets
train_data,test_data,train_values,test_values = train_test_split(X,y,test_size=0.3, random_state=42)

# Iterate over regressors
for regressor,regressor_name in zip(regressors,regressor_names):
    
    regressor.fit(train_data,train_values)
    predicted_values = list(regressor.predict(test_data))
    
    print("-----------------------------------\n")
    print("For Regressor : ",regressor_name)
    print("Mean Absolute Error : ",metrics.mean_absolute_error(list(test_values),predicted_values))
    print("Median Absolute Error : ",metrics.median_absolute_error(list(test_values),predicted_values))
    print( "Mean Squared Error : ",metrics.mean_squared_error(list(test_values),predicted_values))
    print ("R2 Score : ",metrics.r2_score(list(test_values),predicted_values))
    print ("---------------------------------\n")
    
# Perform cross-validation with cross_val_score for each regressor
for regressor, regressor_name in zip(regressors, regressor_names):
    scores = cross_val_score(regressor, X, y, cv=5)  # 5-fold cross-validation
    print(f"{regressor_name} Cross-Validation Scores: {scores}")
    print(f"{regressor_name} Mean Cross-Validation Score: {scores.mean():.2f}\n")
```

**This is solely for the model exploration purposes. Furthermore, there has been minimal hyperparameter tuning so far to keep this as a basic representation. However, tuning can significantly enhance the model's learning capabilities and prevent overfitting.**
