*Consistency & Redundancy Issue:*
- As you can see, there are two columns serving the same function or purpose. Removing one of them is advisable because having duplicates could lead to confusion and inconsistencies in data analysis. This redundancy may cause errors in computations, misinterpretations in reports, and inefficiencies in data management. By eliminating one of the duplicate columns, we ensure clarity, streamline data handling processes, and maintain data integrity throughout our analysis and operations.

- This approach ensures that our data remains consistent and actionable, facilitating more effective decision-making based on accurate and concise information.


First, we address this issue, and subsequently, we proceed to tackle the handling of missing values.


```python
# Removing any of the redundant columns

# Creating a copy of the dataset for modifications
df = titanic.copy()

# Removing Embarked Column
df.drop('embarked', axis=1, inplace=True)

# Display the first few rows of the cleaned DataFrame
display(df.head())
```

So there are 3 cols with null values:

Upon analyzing the above output, since the deck column contains more than 50% null values, it would be advisable to drop it. However, for the Age column, which contains numerical data, we should examine its distribution to decide on an imputation technique. As for the Embark Town column, which is categorical, we will use the mode for imputation.

```python
# Age distribution
sns.displot(df['age'], height=3, aspect=2, kde=True)
plt.show()
```

- The distribution of the Age column does not exhibit skewness, so we will use the mean for imputation.

```python


# Calculate the mean age
mean_age = df['age'].mean()

# Fill missing values in Age with the mean age
df['age'].fillna(mean_age, inplace=True)

# Drop the Cabin column due to high percentage of missing values
df.drop('deck', axis=1, inplace=True)

# Find the most frequent value in Embarked
mode_embarked = df['embark_town'].mode()[0]

# Fill missing values in Embarked with the mode
df['embark_town'].fillna(mode_embarked, inplace=True)

# Display the first few rows of the cleaned DataFrame
display(df.head())
```

**KNN Imputation:**

***Nearest Neighbor Approach:***

- KNNImputer only accepts numerical data and is specifically designed for filling missing values in numerical columns.

- Key Concept: KNN imputation fills missing values based on the values of the nearest neighbors in the feature space.
- Implementation: Requires scaling of numerical features and encoding categorical variables to compute distances accurately.

```python
df_knn_imputed = df.copy()

# Storing original categories before numerical conversion
original_gender_categories = df['Gender'].astype('category').cat.categories
original_product_categories = df['Product Category'].astype('category').cat.categories

# Converting categorical variables to dummy/indicator variables
# df_knn_imputed = pd.get_dummies(df_knn_imputed, drop_first=True)
# Convert categorical variables to numerical categories for KNN imputation
df_knn_imputed['Gender'] = df_knn_imputed['Gender'].astype('category').cat.codes
df_knn_imputed['Product Category'] = df_knn_imputed['Product Category'].astype('category').cat.codes

# Initializing/Creating a KNN imputer with k=3
# For now, I'm using k=3 which refers to the number of nearest neighbors to consider when imputing missing values
knn_imputer = KNNImputer(n_neighbors=3)

# Fitting and transforming the training data for Imputing missing values
# X_imputed = knn_imputer.fit_transform(X)
df_knn_imputed[['Gender', 'Product Category']] = knn_imputer.fit_transform(df_knn_imputed[['Gender', 'Product Category']])   # Imputing only on the columns with missing values

# Converting the imputed NumPy array back to a DataFrame
df_knn_imputed = pd.DataFrame(df_knn_imputed, columns=['Gender', 'Product Category'])

# Converting numerical categories back to categorical after KNN imputation
# df_knn_imputed['Gender'] = pd.Categorical.from_codes(df_knn_imputed['Gender'].round().astype(int), categories=original_gender_categories)
# df_knn_imputed['Product Category'] = pd.Categorical.from_codes(df_knn_imputed['Product Category'].round().astype(int), categories=original_product_categories)

# Displaying the first few rows of the imputed DataFrame
display(df_knn_imputed.sample(3))

# Verifying
print('\nBefore KNN Imputation:\n')
display(df[['Gender','Product Category']].isnull().sum())
print('\nAfter KNN Imputation:')
display(df_knn_imputed.isnull().sum())
```

**MICE (Multiple Imputation by Chained Equations):**

***Iterative Imputation:***
MICE uses a combination of regression models and Bayesian inference to impute missing values. MICE is particularly useful when dealing with complex datasets and multiple types of missing values.

- Sequential Process: MICE imputes missing values one variable at a time, using the current imputation model for other variables.
- Convergence: Multiple iterations may be needed to stabilize imputations and ensure consistency across imputed datasets.

```python
pip install --upgrade miceforest
```

```python
from miceforest import ImputationKernel

df_mice_imputed = df.copy()

# Converting Data Types:

# Converting 'Date' to datetime (for demonstration, we'll exclude it from imputation)
df_mice_imputed['Date'] = pd.to_datetime(df_mice_imputed['Date'])

# Identifying object columns (excluding 'Date' which is handled)
object_columns = df_mice_imputed.select_dtypes(include=['object']).columns.tolist()
# object_columns.remove('Date')  # Remove 'Date' or else apply conversions

# Converting object columns to categorical and then to numeric codes
for column in object_columns:
    df_mice_imputed[column] = df_mice_imputed[column].astype('category').cat.codes

# Initializing ImputationKernel
kernel = ImputationKernel(
    data=df_mice_imputed,           # Data to impute
    datasets=5,        # Number of imputed datasets to create
    random_state=123   # Seed for reproducibility
)

# Performing MICE imputation
kernel.mice(5)       # Perform 5 iterations of MICE

# Getting the imputed data
df_mice_imputed = kernel.complete_data(dataset=0)  # Get the first imputed dataset

# Verifying the Results
print("Missing values before imputation:")
print(df.isnull().sum())
print("Missing values after Mice imputation:")
print(pd.DataFrame(df_mice_imputed).isnull().sum())

# Optionally, save the imputed dataframe to a new CSV file
# pd.DataFrame(df_imputed).to_csv('your_data_imputed.csv', index=False)
```

## **üìè Feature Scaling:**

Normalize or standardize numerical features to ensure they have a similar scale.


**NOTE:**

**Scaling** involves calculating statistics such as mean, standard deviation, minimum, or maximum values of features.

***Feature Scaling Methods***

Three commonly used methods for feature scaling include:


1. **Simple Feature Scaling**:

   Formula: x_new = x_old/x_max

   Description: Scales the feature values to be between 0 and 1 by dividing each value by the maximum value in the feature set.

2. **Standardization (Z-score normalization)**:

   Formula: x_stand = x-mean(x)/std(x)

   Description: Centers the data around 0 (mean) with a standard deviation of 1. This method assumes the data follows a normal distribution.

3. **Normalization (Min-Max scaling)**:

   Formula: x_new = (x-x_min)/(x_max-x_min)

   Description: Transforms the data to a fixed range, typically between 0 and 1, by subtracting the minimum value and dividing by the range (difference between maximum and minimum values).


When applying these scaling methods to training data, it's important to fit the scaling method using only the training data to prevent data leakage from the validation or test sets. This ensures that the scaling parameters (such as mean, standard deviation, minimum, and maximum) are computed solely based on the training data distribution.

> Scaling the input features (X variables) ensures that each feature contributes equally to the analysis

> Scaling the target variable is not as common because it does not affect the relationship between features and the target variable in a meaningful way.

**Using Z-score**
- Z-score helps identify outliers by calculating how many standard deviations away from the mean a data point is.
- Observations with a Z-score greater than 3 or less than -3 are typically considered outliers.


```python
from scipy.stats import zscore

# Calculating Z-score for each feature
z_scores = np.abs(zscore(boston_housing))

# Defining threshold for outliers (typically |Z-score| > 3)
threshold = 3

# Finding indices of outliers
outliers_z = np.where(z_scores > threshold)
print("Indices of outliers detected using Z-score:")
print(list(zip(outliers_z[0], outliers_z[1])))


# Finding no. of outliers
outliers_z = (z_scores > threshold).sum(axis=1)
print("\nNumber of outliers identified by Z-score method:")
print((outliers_z > 0).sum())
```

**Using IQR (Interquartile Range)**
- IQR method defines outliers as those lying outside the whiskers of the boxplot.
-  Observations with values more than 1.5 times the IQR away from Q3 or Q1 are considered outliers

```python
# Calculating IQR (Interquartile Range)
Q1 = boston_housing.quantile(0.25)
Q3 = boston_housing.quantile(0.75)
IQR = Q3 - Q1

# Defining upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Finding no. of outliers
outliers_iqr = ((boston_housing < lower_bound) | (boston_housing > upper_bound)).sum(axis=1)
print("\nNumber of outliers identified by IQR method:")
print((outliers_iqr > 0).sum())

# Finding indices of outliers
outliers_iqr = np.where((boston_housing < lower_bound) | (boston_housing > upper_bound))
print("Indices of outliers detected using IQR:")
print(list(zip(outliers_iqr[0], outliers_iqr[1])))
```

## **üóÇÔ∏è Data Preparation / Transformation:**

**General Transformation Techniques:**

**1. Log Transformation**

* Purpose: Reduce skewness and make the distribution more normal-like. It is used when the data has a large range of values or is exponentially distributed.
* Formula: log(x + 1) or np.log1p(x)
* Effect: Compresses large values and expands small values
* Use cases: Skewed distributions, exponential growth, or when the data has a large range of values

**2. Square Root Transformation**
* Purpose: Reduce skewness and make the distribution more normal-like. It is used when the data is skewed and has a large range of values.
* Formula: sqrt(x)
* Effect: Reduces the effect of extreme values and makes the distribution more symmetric
* Use cases: Skewed distributions, count data, or when the data has a large range of values

**3. Box-Cox Transformation**
* Purpose: Stabilize variance and make the distribution more normal-like. It is used when the data is non-normal and has a large range of values, or when you're unsure which transformation to use.
* Formula: boxcox(x, lambda) Effect: Finds the optimal power transformation to stabilize variance and make the distribution more normal-like
* Use cases: Skewed distributions, non-normal data, or when the data has a large range of values

`Note:` Box-Cox transformations require the data to be strictly positive. Adjust accordingly.

**Key Concepts:**
- Skewness: A measure of how asymmetric a distribution is.
- Normality: A distribution that follows a bell-shaped curve.
- Variance stabilization: Transforming data to have a constant variance across different levels of the predictor variable.
- Power transformation: A family of transformations that raise the data to a power (e.g., log, square root, cube root).

**Correlation Analysis**
- Correlation analysis helps in identifying features that are highly correlated with the target variable or with each other.Features highly correlated with the target variable are usually more important.

- ***Purpose:*** Identify the strength and direction of the linear relationship between features.
- ***Why Use It:*** Simplifies feature selection and detects multicollinearity.

- **1. Correlation Magnitude**

> - High Positive Correlation (e.g., |correlation| > 0.5)
> - Moderate Positive Correlation (e.g., 0.3 < |correlation| ‚â§ 0.5)
> - Moderate to Low Positive Correlation (e.g., 0.2 < |correlation| ‚â§ 0.3)
> - Low Positive Correlation (e.g., 0.1 < |correlation| ‚â§ 0.2)
> - Very Low Positive Correlation (e.g., |correlation| ‚â§ 0.1)
> - Negative Correlation (e.g., |correlation| > 0.3)


```python
# Computing correlation matrix
corr_matrix = diabetes.corr()

# Plotting heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix', fontsize =16, fontweight='bold', pad=10)
plt.show()

# Displaying correlation with the target variable
print("Correlation with target:")
print(corr_matrix['target'].sort_values(ascending=False))
```

**Interpreting Correlation Coefficients:**
This matrix is showing correlation from strongest to lowest.

* High Correlation:

  Including features with high positive correlations (e.g., bmi, s5) as they are strong predictors of the target.

* Moderate Correlation:

  Considered including features with moderate positive correlations (e.g., bp, s4) for additional predictive power.

* Low Correlation:

  Evaluating features with low correlations (e.g., age, s2). They may offer some value but are less impactful.

* Negative Correlation:

  Including features with moderate negative correlations (e.g., s3) if they provide valuable insights into the target variable's behavior.

**Mutual Information**

- Mutual information measures the dependency between features and the target variable. It can capture non-linear relationships and identifies informative features.. Higher mutual information indicates a stronger relationship.

- Note:
  - High Mutual Information (e.g., >0.1): Include these features in your model as they are most informative.
  - Moderate Mutual Information (e.g., 0.05 - 0.1): Consider these features, but they might be less critical compared to high-scoring features.
  - Low Mutual Information (e.g., <0.05): These features might be excluded from the model as they provide minimal additional information about the target variable.

```python
from sklearn.feature_selection import mutual_info_regression

# Computing mutual information
mi = mutual_info_regression(diabetes.drop(columns='target'), diabetes['target'])

# Creating a DataFrame to display mutual information scores
mi_df = pd.DataFrame({'Feature': diabetes.columns[:-1], 'Mutual Information': mi})  # [:-1] --> excludes target col
mi_df = mi_df.sort_values(by='Mutual Information', ascending=False)

print("Mutual Information Scores:")
print(mi_df)
```

**MI Analysis and Conclusion:**

- bmi and s5 should be prioritized as they have the highest mutual information scores, indicating they are the most valuable features for predicting the target variable.

- Features such as s6 and s4 should be included as they provide useful information, though they are not as critical as the top features.

- Features with lower mutual information scores (s3, s1, bp) should be evaluated for their added value. You may consider including them if necessary, but be aware that their contribution might be less significant.

- sex, s2, and age have very low or no mutual information with the target and can likely be excluded from the model to simplify it and focus on features with more substantial contributions.




---


**Recursive Feature Elimination (RFE)**
- RFE selects features by recursively considering smaller and smaller sets of features. It uses a model to evaluate the importance of each feature because itdentifies the best subset of features, useful with model-based selection.

- Use RFE to recursively remove less important features based on a chosen model's importance.


```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Separating/Splitting features and target
X = diabetes.drop(columns=['target'])
y = diabetes['target']

# Initializing the model
model = LinearRegression()

# Initializing RFE
rfe = RFE(model, n_features_to_select=5)  # Select top 5 features
rfe = rfe.fit(X, y)

# Creating DataFrame to display RFE results
rfe_df = pd.DataFrame({'Feature': diabetes.columns[:-1], 'RFE Support': rfe.support_, 'RFE Ranking': rfe.ranking_})
rfe_df = rfe_df.sort_values(by='RFE Ranking')

# Displaying RFE results
print("RFE Results:")
print(rfe_df)



# Getting selected features
selected_features = X.columns[rfe.support_]
print("\nSelected Features by RFE:")
print(selected_features)
```

**RFE Analysis**
- The features with a ranking of 1 should be prioritized, while those with higher rankings may be considered for exclusion, leading to a more streamlined and effective model.





## **üìä‚öôÔ∏è Dealing with Imbalanced Data**

### Credit Card Fraud Detection Dataset


```python
creditcardfraud.head(3)
```

```python
creditcardfraud.isnull().sum()
```

```python
# Checking the distribution of target variable
print("\nClass Frequency:")
print(creditcardfraud['Class'].value_counts())
```


As observed, the dataset is highly imbalanced with 284,315 instances of class 0 and only 492 instances of class 1. This significant disparity can lead to biased outcomes in model predictions.




```python
# Handling imbalanced data in using techniques like SMOTE, ADASYN, and Undersampling:

from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import train_test_split
```

```python
# Splitting the dataset into features and target
X = creditcardfraud.drop('Class', axis=1)
y = creditcardfraud['Class']

# Splitting the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

**SMOTE (Synthetic Minority Over-sampling Technique)**
- SMOTE creates synthetic samples of the minority class by interpolating between existing minority class instances.
- This helps to increase the size of the minority class, making the dataset more balanced.

```python
# Applying SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
# X_resampled, y_resampled = smote.fit_resample(X, y)
X_resampled_smote, y_resampled_smote = smote.fit_resample(X_train, y_train)
```

```python
# Converting the resampled arrays back to DataFrames
X_smote_df = pd.DataFrame(X_resampled_smote, columns=X_train.columns)
y_smote_df = pd.DataFrame(y_resampled_smote, columns=['Class'])

# Combining X and y to view the complete DataFrame
smote_df = pd.concat([X_smote_df, y_smote_df], axis=1)

smote_df['Class'].value_counts()
```

**ADASYN (Adaptive Synthetic Sampling)**
- ADASYN is an extension of SMOTE that adaptively generates synthetic samples based on the density of the minority class instances. It generates samples with a focus on harder-to-learn examples.

```python
# Applying ADASYN
from imblearn.over_sampling import ADASYN

adasyn = ADASYN(random_state=42)
X_resampled_adasyn, y_resampled_adasyn = adasyn.fit_resample(X, y)
```

```python
# Converting the resampled arrays back to DataFrames
X_smote_df = pd.DataFrame(X_resampled_adasyn, columns=X_train.columns)
y_smote_df = pd.DataFrame(y_resampled_adasyn, columns=['Class'])

# Combining X and y to view the complete DataFrame
smote_df = pd.concat([X_smote_df, y_smote_df], axis=1)

smote_df['Class'].value_counts()
```

**Undersampling**
- Undersampling involves reducing the size of the majority class to match the size of the minority class.
- Randomly remove samples from the majority class.

```python
# Applying Under Sampling -- RandomUnderSampler

from imblearn.under_sampling import RandomUnderSampler

undersampler = RandomUnderSampler(random_state=42)
X_resampled_undersample, y_resampled_undersample = undersampler.fit_resample(X, y)
```

```python
# Converting the resampled arrays back to DataFrames
X_smote_df = pd.DataFrame(X_resampled_undersample, columns=X_train.columns)
y_smote_df = pd.DataFrame(y_resampled_undersample, columns=['Class'])

# Combining X and y to view the complete DataFrame
smote_df = pd.concat([X_smote_df, y_smote_df], axis=1)

smote_df['Class'].value_counts()
```

```python
# Checking the class distribution after resampling
print("\nClass Frequencies after SMOTE:")
print(pd.Series(y_resampled_smote).value_counts())

print("\nClass Frequencies after ADASYN:")
print(pd.Series(y_resampled_adasyn).value_counts())

print("\nClass Frequencies after Undersampling:")
print(pd.Series(y_resampled_undersample).value_counts())
```

**Model Evaluation**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

def evaluate_model(X_train, y_train, X_test, y_test, model_name):

    # Initializing and train the Logistic Regression model
    model = LogisticRegression(solver='liblinear')
    model.fit(X_train, y_train)

    # Making predictions on the test set
    y_pred = model.predict(X_test)

    # Printing the classification report & confusion matrix
    print(f"\n{model_name} - Classification Report:")
    print(classification_report(y_test, y_pred))
    print(f"\n{model_name} - Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
```

```python
display(evaluate_model(X_resampled_smote, y_resampled_smote, X_test, y_test, "SMOTE"))
display(evaluate_model(X_resampled_adasyn, y_resampled_adasyn, X_test, y_test, "ADASYN"))
display(evaluate_model(X_resampled_undersample, y_resampled_undersample, X_test, y_test, "Undersampling"))
```

In this case, SMOTE is the best approach because it achieved 99% accuracy and resulted in a lower number of True Negative values compared to Undersampling and ADASYN.
