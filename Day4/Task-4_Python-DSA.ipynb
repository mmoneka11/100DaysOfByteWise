```python
#1

def quick_sort(array):
    """
    Sorts an array in-place using the QuickSort algorithm.
    
    Args:
    - array: List of elements to be sorted
    
    Returns:
    - Sorted list of elements
    """
    if len(array) < 2:
        return array  # Base case: If array has 0 or 1 elements, it is already sorted
    
    else:
        pivot = array[0]  # Choose the first element as the pivot
        # Partitioning step
        less = [i for i in array[1:] if i <= pivot]  # Elements less than or equal to pivot
        greater = [i for i in array[1:] if i > pivot]  # Elements greater than pivot
        
        # Recursively sort the partitions
        sorted_less = quick_sort(less)
        sorted_greater = quick_sort(greater)
        
        # Combine sorted partitions with the pivot and return
        return sorted_less + [pivot] + sorted_greater
    
#         return quickSort(less) + [pivot] + quickSort(greater)


if __name__ == '__main__':
    # Example usage
    array = [3, 6, 8, 10, 1, 2, 1]
    print(f"Original Array: {array}")
    
    # Sort the array using QuickSort
    sorted_array = quick_sort(array)
    
    # Print the sorted array
    print(f"Sorted Array: {sorted_array}")
```

```python
#2

def knapsack(weights, values, capacity):
    """
    Solves the 0/1 Knapsack problem using dynamic programming.

    Args:
    - weights: List of weights of items
    - values: List of values of items
    - capacity: Capacity of the knapsack

    Returns:
    - Maximum value that can be obtained with the given capacity
    """
    n = len(weights)
    
    # Initialize a 2D list (dp) to store maximum values for different capacities and items
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]
    
    # Fill the dp table
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i-1] <= w:
                # If the current item can be included, choose the maximum of including it or excluding it
                dp[i][w] = max(values[i-1] + dp[i-1][w-weights[i-1]], dp[i-1][w])
            else:
                # If the current item cannot be included, carry forward the previous value
                dp[i][w] = dp[i-1][w]
    
    # The bottom-right cell contains the maximum value achievable with the given capacity
    return dp[n][capacity]


if __name__ == '__main__':
    # Example usage
    weights = [1, 3, 4, 5]
    values = [1, 4, 5, 7]
    capacity = 7

    # Solve the knapsack problem and print the maximum value
    max_value = knapsack(weights, values, capacity)
    print(f"Maximum value that can be obtained with capacity {capacity} is: {max_value}")
```

```python
#3

from collections import defaultdict

class Graph:
    def __init__(self):
        self.adjacency_list = defaultdict(list)

    def add_edge(self, start, end):
        """
        Adds an edge between two nodes in the graph.
        
        Args:
        - start: Starting node of the edge
        - end: Ending node of the edge
        """
        self.adjacency_list[start].append(end)

    def breadth_first_search(self, start_node):
        """
        Performs Breadth-First Search (BFS) starting from a given node.
        
        Args:
        - start_node: Node to start BFS from
        
        Returns:
        - List of nodes visited in BFS order
        """
        visited = [False] * (max(self.adjacency_list) + 1)
        queue = []
        result = []

        queue.append(start_node)
        visited[start_node] = True

        while queue:
            current_node = queue.pop(0)
            result.append(current_node)

            for neighbor in self.adjacency_list[current_node]:
                if not visited[neighbor]:
                    queue.append(neighbor)
                    visited[neighbor] = True

        return result

    def depth_first_search(self, start_node):
        """
        Performs Depth-First Search (DFS) starting from a given node.
        
        Args:
        - start_node: Node to start DFS from
        
        Returns:
        - List of nodes visited in DFS order
        """
        result = []

        def dfs_recursive(node, visited):
            """
            Helper function to perform DFS recursively.
            
            Args:
            - node: Current node in DFS traversal
            - visited: List to track visited nodes
            """
            result.append(node)
            visited[node] = True

            for neighbor in self.adjacency_list[node]:
                if not visited[neighbor]:
                    dfs_recursive(neighbor, visited)

        visited = [False] * (max(self.adjacency_list) + 1)
        dfs_recursive(start_node, visited)

        return result


if __name__ == "__main__":
    # Example usage of the Graph class
    g = Graph()

    # Adding edges to the graph
    g.add_edge(0, 1)
    g.add_edge(0, 2)
    g.add_edge(1, 2)
    g.add_edge(2, 0)
    g.add_edge(2, 3)
    g.add_edge(3, 3)

    # Performing BFS from node 2
    bfs_result = g.breadth_first_search(2)
    print("Breadth-First Search starting from node 2:")
    print(bfs_result)
    print()

    # Performing DFS from node 2
    dfs_result = g.depth_first_search(2)
    print("Depth-First Search starting from node 2:")
    print(dfs_result)
```

```python
#4

def dijkstra(graph, start_node):
    """
    Computes the shortest paths from a start node to all other nodes in a graph using Dijkstra's algorithm.

    Args:
    - graph: Dictionary representing the graph with nodes as keys and their neighbors with weights as values
    - start_node: Starting node for Dijkstra's algorithm

    Returns:
    - Dictionary where keys are nodes and values are the shortest distances from the start_node to each node
    """
    # Initialize distances dictionary with infinity for all nodes except the start node
    distances = {node: float('inf') for node in graph}
    distances[start_node] = 0

    # Initialize priority queue with tuple (distance, node)
    priority_queue = [(0, start_node)]

    # Set to keep track of visited nodes
    visited = set()

    while priority_queue:
        # Sort priority queue by distance and extract node with smallest distance
        priority_queue.sort()
        current_distance, current_node = priority_queue.pop(0)

        # Skip processing if node has already been visited
        if current_node in visited:
            continue
        
        # Mark current node as visited
        visited.add(current_node)

        # Update distances to neighbors if shorter path is found
        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight
            
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                priority_queue.append((distance, neighbor))

    return distances

if __name__ == '__main__':
    # Example usage
    graph = {
        'A': {'B': 1, 'C': 4},
        'B': {'C': 2, 'D': 5},
        'C': {'D': 1},
        'D': {}
    }

    start_node = 'A'
    shortest_paths = dijkstra(graph, start_node)
    print("Shortest paths from node", start_node)
    print(shortest_paths)
```
